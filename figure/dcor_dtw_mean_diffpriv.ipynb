```python
import pandas as pd
import numpy as np

import os
```

```python
from google.colab import drive

drive.mount('/content/gdrive')
root_path = 'gdrive/My Drive/Colab Notebooks'
```

```python
dcorlist = []
dtwlist = []
fnamelist = ['mean', 'mean_diffpriv_ep10', 'mean_diffpriv_ep7', 'mean_diffpriv_ep5', 'mean_diffpriv_ep3', 'mean_diffpriv_ep1']
fnamelist2 = ['mean', 'mean_diffpriv_ep10', 'mean_diffpriv_ep7', 'mean_diffpriv_ep5', 'mean_diffpriv_ep3', 'mean_diffpriv_ep1']
for fname in fnamelist:
    a = np.genfromtxt(os.path.join(root_path, 'dtw_dcor', 'conv2_dcor_{}.csv'.format(fname)), delimiter=',')
    dcorlist.append(a)
for fname in fnamelist2:
    b = np.genfromtxt(os.path.join(root_path, 'dtw_dcor', 'conv2_dtw_{}.csv'.format(fname)), delimiter=',')
    dtwlist.append(b)
```

```python
np.savetxt(os.path.join(root_path, 'csv', 'dcor_mean_diffpriv.csv'), np.asarray(dcorlist), '%.2f', delimiter=',')
np.savetxt(os.path.join(root_path, 'csv', 'dtw_mean_diffpriv.csv'), np.asarray(dtwlist), '%.2f', delimiter=',')
```

```python
import matplotlib.pyplot as plt

num_plots = 16

dcor_ylist = []
for i in range(16):
    tmp = []
    for dl in dcorlist:
        tmp.append(dl[i])
    dcor_ylist.append(tmp)
plt.figure()
plt.gca().set_prop_cycle(plt.cycler('color', plt.cm.YlOrRd(np.linspace(1, 0.3, num_plots))))
for j in range(16):
    plt.plot(['None', 10, 7, 5, 3, 1], dcor_ylist[j], 'o-', lw=1)
    
plt.xlabel('Epsilon', size=12)
plt.ylabel('Distance Correlation', size=12)
plt.tick_params(labelsize=12)
plt.ylim(0,1)
plt.grid(alpha=0.5)
# plt.savefig(os.path.join(root_path, 'img', 'dcor_mean_diffpriv.pdf'), bbox_inches='tight')
plt.show()
```

```python
dtw_ylist = []
for i in range(16):
    tmp = []
    for dl in dtwlist:
        tmp.append(dl[i])
    dtw_ylist.append(tmp)
plt.figure()
plt.gca().set_prop_cycle(plt.cycler('color', plt.cm.YlOrRd(np.linspace(1, 0.3, num_plots))))
for j in range(16):
    plt.plot(['None', 10, 7, 5, 3, 1], dtw_ylist[j], 'o-', lw=1)
    
plt.xlabel('Epsilon', size=12)
plt.ylabel('DTW', size=12)
plt.tick_params(labelsize=12)
plt.grid(alpha=0.5)
# plt.savefig(os.path.join(root_path, 'img', 'dtw_mean_diffpriv.pdf'), bbox_inches='tight')
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(14,4))
c = plt.cycler('color', plt.cm.YlOrRd(np.linspace(1, 0.3, num_plots)))

fig.text(0.085, 0, '(a)', size=20)
fig.text(0.505, 0, '(b)', size=20)

ax[0].set_prop_cycle(c)
ax[1].set_prop_cycle(c)

ax[0].set_xlabel('Epsilon', size=16)
ax[0].set_ylabel('Distance Correlation', size=16)
ax[0].tick_params(labelsize=16)
ax[0].grid(alpha=0.5)
for j in range(16):
    ax[0].plot(['None', 10, 7, 5, 3, 1], dcor_ylist[j], 'o-', lw=1)

ax[1].set_xlabel('Epsilon', size=16)
ax[1].set_ylabel('DTW', size=16)
ax[1].tick_params(labelsize=16)
ax[1].grid(alpha=0.5)
for j in range(16):
    ax[1].plot(['None', 10, 7, 5, 3, 1], dtw_ylist[j], 'o-', lw=1)

fig.savefig(os.path.join(root_path, 'img', 'dcor_dtw_mean_diffpriv.pdf'), bbox_inches='tight')
```
